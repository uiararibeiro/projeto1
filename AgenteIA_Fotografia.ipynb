```python
from google.adk.agents import Agent
from google.adk.runners import Runner
from google.adk.sessions import InMemorySessionService
from google.adk.tools import google_search
from google.genai import types  # Para criar conteÃºdos (Content e Part)
from datetime import date
import textwrap # Para formatar melhor a saÃ­da de texto
from IPython.display import display, Markdown # Para exibir texto formatado no Colab
import requests # Para fazer requisiÃ§Ãµes HTTP
import warnings

warnings.filterwarnings("ignore")
```

```python
# FunÃ§Ã£o auxiliar que envia uma mensagem para um agente via Runner e retorna a resposta final
def call_agent(agent: Agent, message_text: str) -> str:
    # Cria um serviÃ§o de sessÃ£o em memÃ³ria
    session_service = InMemorySessionService()
    # Cria uma nova sessÃ£o (vocÃª pode personalizar os IDs conforme necessÃ¡rio)
    session = session_service.create_session(app_name=agent.name, user_id="user1", session_id="session1")
    # Cria um Runner para o agente
    runner = Runner(agent=agent, app_name=agent.name, session_service=session_service)
    # Cria o conteÃºdo da mensagem de entrada
    content = types.Content(role="user", parts=[types.Part(text=message_text)])

    final_response = ""
    # Itera assincronamente pelos eventos retornados durante a execuÃ§Ã£o do agente
    for event in runner.run(user_id="user1", session_id="session1", new_message=content):
        if event.is_final_response():
          for part in event.content.parts:
            if part.text is not None:
              final_response += part.text
              final_response += "\n"
    return final_response
```

```python
# FunÃ§Ã£o auxiliar para exibir texto formatado em Markdown no Colab
def to_markdown(text):
  text = text.replace('â€¢', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

```python
##########################################
# --- Agente 1: Buscador de NotÃ­cias --- #
##########################################
def agente_buscador(topico, data_de_hoje):

    buscador = Agent(
        name="agente_buscador",
        model="gemini-2.0-flash",
        instruction="""
        VocÃª Ã© um assistente de pesquisa. A sua tarefa Ã© usar a ferramenta de busca do google (google_search)
        para recuperar as Ãºltimas notÃ­cias de lanÃ§amentos muito relevantes sobre o tÃ³pico abaixo.
        Foque em no mÃ¡ximo 5 lanÃ§amentos relevantes, com base na quantidade e entusiasmo das notÃ­cias sobre ele.
        Se um tema tiver poucas notÃ­cias ou reaÃ§Ãµes entusiasmadas, Ã© possÃ­vel que ele nÃ£o seja tÃ£o relevante assim
        e pode ser substituÃ­do por outro que tenha mais.
        Esses lanÃ§amentos relevantes devem ser atuais, de no mÃ¡ximo um mÃªs antes da data de hoje.
        """,
        description="Agente que busca informaÃ§Ãµes no Google",
        tools=[google_search]
    )

    entrada_do_agente_buscador = f"TÃ³pico: {topico}\nData de hoje: {data_de_hoje}"

    lancamentos = call_agent(buscador, entrada_do_agente_buscador)
    return lancamentos
```

```python
################################################
# --- Agente 2: Planejador de posts --- #
################################################
def agente_planejador(topico, lancamentos_buscados):
    planejador = Agent(
        name="agente_planejador",
        model="gemini-2.0-flash",
        # Inserir as instruÃ§Ãµes do Agente Planejador #################################################
        instruction="""
        VocÃª Ã© um planejador de conteÃºdo, especialista em redes sociais. Com base na lista de
        lanÃ§amentos mais recentes e relevantes buscador, vocÃª deve:
        usar a ferramenta de busca do Google (google_search) para criar um plano sobre
        quais sÃ£o os pontos mais relevantes que poderÃ­amos abordar em um post sobre
        cada um deles. VocÃª tambÃ©m pode usar o (google_search) para encontrar mais
        informaÃ§Ãµes sobre os temas e aprofundar.
        Ao final, vocÃª irÃ¡ escolher o tema mais relevante entre eles com base nas suas pesquisas
        e retornar esse tema, seus pontos mais relevantes, e um plano com os assuntos
        a serem abordados no post que serÃ¡ escrito posteriormente.
        """,
        description="Agente que planeja posts",
        tools=[google_search]
    )

    entrada_do_agente_planejador = f"TÃ³pico:{topico}\nLanÃ§amentos buscados: {lancamentos_buscados}"
    # Executa o agente
    plano_do_post = call_agent(planejador, entrada_do_agente_planejador)
    return plano_do_post
```

```python
######################################
# --- Agente 3: Redator do Post --- #
######################################
def agente_redator(topico, plano_de_post):
    redator = Agent(
        name="agente_redator",
        model="gemini-2.0-flash5",
        instruction="""
            VocÃª Ã© um Redator Criativo especializado em criar posts virais para redes sociais.
            VocÃª escreve posts para a empresa ValFotosInstax, a maior escola online de tecnologia do Brasil.
            Utilize o tema fornecido no plano de post e os pontos mais relevantes fornecidos e, com base nisso,
            escreva um rascunho de post para Instagram sobre o tema indicado.
            O post deve ser engajador, informativo, com linguagem simples e incluir 2 a 4 hashtags no final.
            """,
        description="Agente redator de posts engajadores para Instagram"
    )
    entrada_do_agente_redator = f"TÃ³pico: {topico}\nPlano de post: {plano_de_post}"
    # Executa o agente
    rascunho = call_agent(redator, entrada_do_agente_redator)
    return rascunho
```

```python
##########################################
# --- Agente 4: Revisor de Qualidade --- #
##########################################
def agente_revisor(topico, rascunho_gerado):
    revisor = Agent(
        name="agente_revisor",
        model="gemini-2.0-flash",
        instruction="""
            VocÃª Ã© um Editor e Revisor de ConteÃºdo meticuloso, especializado em posts para redes sociais, com foco no Instagram.
            Por ter um pÃºblico adulto, entre 18 e 60 anos, use um tom de escrita adequado.
            Revise o rascunho de post de Instagram abaixo sobre o tÃ³pico indicado, verificando clareza, concisÃ£o, correÃ§Ã£o e tom.
            Se o rascunho estiver bom, responda apenas 'O rascunho estÃ¡ Ã³timo e pronto para publicar!'.
            Caso haja problemas, aponte-os e sugira melhorias.
            """,
        description="Agente revisor de post para redes sociais."
    )
    entrada_do_agente_revisor = f"TÃ³pico: {topico}\nRascunho: {rascunho_gerado}"
    # Executa o agente
    texto_revisado = call_agent(revisor, entrada_do_agente_revisor)
    return texto_revisado
```

```python
data_de_hoje = date.today().strftime("%d/%m/%Y")

print("ğŸš€ Iniciando o Sistema de CriaÃ§Ã£o de Posts para Instagram com 4 Agentes ğŸš€")

# --- Obter o TÃ³pico do UsuÃ¡rio ---
topico = input("â“ Por favor, digite o TÃ“PICO sobre o qual vocÃª quer criar o post de tendÃªncias: ")

# Inserir lÃ³gica do sistema de agentes ################################################
if not topico:
    print("VocÃª esqueceu de digitar o tÃ³pico!")
else:
    print(f"Maravilha! Vamos entÃ£o criar o post sobre novidades em {topico}")

    lancamentos_buscados = agente_buscador(topico, data_de_hoje)
    print("\n--- ğŸ“ Resultado do Agente 1 (Buscador) ---\n")
    display(to_markdown(lancamentos_buscados))
    print("--------------------------------------------------------------")

    plano_de_post = agente_planejador(topico, lancamentos_buscados)
    print("\n--- ğŸ“ Resultado do Agente 2 (Planejador) ---\n")
    display(to_markdown(plano_de_post))
    print("--------------------------------------------------------------")

    rascunho_de_post = agente_redator(topico, plano_de_post)
    print("\n--- ğŸ“ Resultado do Agente 3 (Redator) ---\n")
    display(to_markdown(rascunho_de_post))
    print("--------------------------------------------------------------")

    post_final = agente_revisor(topico, rascunho_de_post)
    print("\n--- ğŸ“ Resultado do Agente 4 (Revisor) ---\n")
    display(to_markdown(post_final))
    print("--------------------------------------------------------------")
```
